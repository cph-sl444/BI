```python
# read the Json file from your data folder into a data frame
df = pd.read_json('Sample2.json')
```

```python
# see the size
df.shape
```

```python
print(df)
```

```python
# count the missing values
df.isnull().sum()
```

```python
# Flatten the nested structure
flat_data = {}
for category, items in df['shoppingList'].items():
    for item in items:
        for key, value in item.items():
            flat_data.setdefault(key, []).append(value)
```

```python
df = pd.DataFrame(flat_data)
```

```python
print(df)
```

```python
# mean
# np.mean(df[['age']])
np.mean(df.quantity)
```

```python
# box plot
# here we see the median25th and 75th percentiles, the range, and the outliers
df.quantity.plot.box()
```

```python
# enhancement to the boxplot
# shos the dencity estimates
sns.violinplot(data=df, x='item', y='quantity', inner="quartile", color="white")

# Add labels and title
plt.xlabel('Item')
plt.ylabel('Quantity')
plt.title('Violin Plot with Density Estimates')

# Show the plot
plt.show()
```
